#### Using Webscraping with Selenium
Run the below python code to open the Safari web browser automatically.

```python
from selenium import webdriver
import time
browser = webdriver.Safari(executable_path = '/usr/bin/safaridriver')    
browser.get('https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000&punto=indice')
#time.sleep(10)
#browser.quit()
```

In this case, I will use the xpath to get the syntax oh the html page.
For example, I want to know the syntax of the 'ACCIONA'. When I copy its xpath , this is shown:

**//*[@id="ctl00_Contenido_tblAcciones"]/tbody/tr[2]/td[1]/a**

So, I will use the function 'find_elements_by_xpath' from Selenium to confirm the value of the cell. I also put '[0]' because it is the first row element.

```python
browser.find_elements_by_xpath('//*[@id="ctl00_Contenido_tblAcciones"]/tbody/tr[2]/td[1]/a')[0].text
```

```python
# Create a dataframe for Selenium

Nombre = []
Ult= []
Dif= []
Max= []
Min= []
Volumen= []
Efectivo = []
Fecha= []
Hora= []



# Looping through the  36 rows of the webpage table,
for i in range(2, 37):
    
    
    
    Nombre.append(browser.find_elements_by_xpath('//*[@id="ctl00_Contenido_tblAcciones"]/tbody/tr[' + str(i) + ']/td[1]/a')[0].text)
    Ult.append(browser.find_elements_by_xpath('//*[@id="ctl00_Contenido_tblAcciones"]/tbody/tr[' + str(i) + ']/td[2]')[0].text)
    Dif.append(browser.find_elements_by_xpath('//*[@id="ctl00_Contenido_tblAcciones"]/tbody/tr[' + str(i) + ']/td[3]')[0].text)
    Max.append(browser.find_elements_by_xpath('//*[@id="ctl00_Contenido_tblAcciones"]/tbody/tr[' + str(i) + ']/td[4]')[0].text)
    Min.append(browser.find_elements_by_xpath('//*[@id="ctl00_Contenido_tblAcciones"]/tbody/tr[' + str(i) + ']/td[5]')[0].text)
    Volumen.append(browser.find_elements_by_xpath('//*[@id="ctl00_Contenido_tblAcciones"]/tbody/tr[' + str(i) + ']/td[6]')[0].text)
    Efectivo.append(browser.find_elements_by_xpath('//*[@id="ctl00_Contenido_tblAcciones"]/tbody/tr[' + str(i) + ']/td[7]')[0].text)
    Fecha.append(browser.find_elements_by_xpath('//*[@id="ctl00_Contenido_tblAcciones"]/tbody/tr[' + str(i) + ']/td[8]')[0].text)
    Hora.append(browser.find_elements_by_xpath('//*[@id="ctl00_Contenido_tblAcciones"]/tbody/tr[' + str(i) + ']/td[9]')[0].text)
    
    
# After the loop is finished, I create the DataFrame.
df_selenium = pd.DataFrame({'Nombre': Nombre, 
                            'Ultimo' : Ult, 
                            'Dif'  : Dif, 
                            'Max'   : Max, 
                            'Min'   : Min,
                            'Volumen': Volumen,
                            'Efectivo' :Efectivo,
                            'Fecha' :Fecha,
                            'Hora' : Hora})
```

```python
#import libraries
import scrapy
from scrapy.crawler import CrawlerProcess
import bs4 as bs
import urllib
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError

 #opens a file to write
f = open('op.txt','w')

#Exception Handling
try:                           
    html = urlopen("https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250")
except HTTPError as e:
    print("The server returned a HTTP error")
except URLError as e:
    print("The server could not be found!")
else:
    soup=bs.BeautifulSoup(html,'lxml')
    print(soup.prettify(),file=f) #writes the source code to a text file
    f = open("op.txt", "r")
    print(f.read()) #reads the file and prints the source code
```

The following pipeline stores all scraped itemsinto a single items.jl file, containing one item per line serialized in JSON format.

```python
# write items to a JSON file
import json

class JsonWriterPipeline(object):

    def open_spider(self, spider):
        self.file = open('imdb.jl', 'w')
        self.file1 = open('imdb.json', 'w')

    def close_spider(self, spider):
        self.file.close()
        self.file1.close()

    def process_item(self, item, spider):
        line = json.dumps(dict(item)) + "\n"
        self.file.write(line)
        self.file1.write(line)
        return item
```

Next, the logger is created using the Spider’s name (IMDB).

```python
# Logging from Spiders
import logging

class IMDB(scrapy.Spider):
    name = "imdb"
    start_urls = ['https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250']
    custom_settings = {
        'LOG_LEVEL': logging.WARNING,
        'ITEM_PIPELINES': {'__main__.JsonWriterPipeline': 1}, # Used for pipeline 1
        'FEED_FORMAT':'json',                                 # Used for pipeline 2
        'FEED_URI': 'imdb.json'                        # Used for pipeline 2
    }
    
    def parse(self, response):
        for db in response.css('tbody.lister-list tr'):
            yield {
                'position': db.css('td.titleColumn::text').extract_first(),
                'name': db.css('td.titleColumn a::text').extract_first(),
                'year': db.css('td.titleColumn span.secondaryInfo::text').extract_first(),
                'artists':db.css("td.titleColumn a::attr(title)").extract_first(),
                'rating':db.css('td strong::text').extract_first(),                
                'poster':db.css("td.posterColumn img::attr(src)").extract_first(),
                'moive_link':db.css("td.posterColumn a::attr(href)").extract_first(),
            }
```

Scrapy is built on top of the Twisted asynchronous networking library, so, to start a Twisted reactor, configuring the logging and setting shutdown handlers, let's use CrawlerProcess.

```python
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})

process.crawl(IMDB)
process.start()
```
